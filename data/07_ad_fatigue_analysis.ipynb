# Ad Fatigue and User Experience Analysis


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load the cleaned dataset
try:
    df = pd.read_csv('../processed/survey_cleaned.csv')
except FileNotFoundError:
    df = pd.read_csv('processed/survey_cleaned.csv')
```

### 16. Advertisement Frequency Distribution
Bar chart showing how often users see ads.


```
plt.figure()
order = ['Rarely', 'Sometimes', 'Often', 'Always']
order = [x for x in order if x in df['ad_frequency'].unique()]
if not order: order = df['ad_frequency'].value_counts().index
sns.countplot(data=df, x='ad_frequency', order=order, palette='viridis')
plt.title("Advertisement Frequency Distribution")
plt.xlabel("Ad Frequency")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### 17. Ads Interrupt Usage
Bar chart showing user responses.


```
plt.figure()
order = ['Strongly disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly agree']
actual_order = [x for x in order if x in df['ads_interrupt_usage'].unique()]
sns.countplot(data=df, x='ads_interrupt_usage', order=actual_order, palette='coolwarm')
plt.title("Ads Interrupt Usage")
plt.xlabel("Response")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### 18. Ads Cause Frustration
Bar chart showing frustration levels.


```
plt.figure()
actual_order = [x for x in order if x in df['ads_cause_frustration'].unique()]
sns.countplot(data=df, x='ads_cause_frustration', order=actual_order, palette='coolwarm')
plt.title("Ads Cause Frustration")
plt.xlabel("Frustration Level")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### 19. Ads Reduce Enjoyment
Bar chart showing impact on user experience.


```
plt.figure()
actual_order = [x for x in order if x in df['ads_reduce_enjoyment'].unique()]
sns.countplot(data=df, x='ads_reduce_enjoyment', order=actual_order, palette='coolwarm')
plt.title("Ads Reduce Enjoyment")
plt.xlabel("Impact on Enjoyment")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### 20. Ads Close the App
Bar chart showing how often users exit apps due to ads.


```
plt.figure()
actual_order = [x for x in order if x in df['ads_close_app'].unique()]
sns.countplot(data=df, x='ads_close_app', order=actual_order, palette='coolwarm')
plt.title("Ads Cause Users to Close the App")
plt.xlabel("Response")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### 21. Most Annoying Ad Timing
Bar chart identifying when ads are most irritating.


```
plt.figure()
sns.countplot(data=df, x='most_annoying_ad_time', palette='magma', order=df['most_annoying_ad_time'].value_counts().index)
plt.title("Most Annoying Ad Timing")
plt.xlabel("Time of Day")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

### 22. Preferred Ad Format
Pie chart showing preferred (tolerated) advertisement formats.


```
plt.figure(figsize=(8, 8))
format_counts = df['tolerated_ad_format'].value_counts()
plt.pie(format_counts, labels=format_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Preferred / Tolerated Ad Format")
plt.show()
```

### 23. Tolerated Ad Format
Bar chart showing acceptable ad formats.


```
plt.figure()
sns.countplot(data=df, y='tolerated_ad_format', palette='Set2', order=df['tolerated_ad_format'].value_counts().index)
plt.title("Tolerated Ad Formats")
plt.xlabel("Number of Users")
plt.ylabel("Ad Format")
plt.show()
```

### 24. Ad Frequency vs Frustration
Box plot showing relationship between ad exposure and frustration.


```
plt.figure()
freq_order = ['Rarely', 'Sometimes', 'Often', 'Always']
freq_order = [x for x in freq_order if x in df['ad_frequency'].unique()]
sns.boxplot(data=df, x='ad_frequency', y='ads_cause_frustration_score', palette='Set3', order=freq_order)
plt.title("Ad Frequency vs Frustration Score")
plt.xlabel("Ad Frequency")
plt.ylabel("Frustration Score (1-5)")
plt.show()
```

### 25. Age Group vs Ad Frustration
Stacked bar chart analyzing frustration across age groups.


```
crosstab_data = pd.crosstab(df['age_group'], df['ads_cause_frustration'])
col_order = ['Strongly disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly agree']
actual_col_order = [x for x in col_order if x in crosstab_data.columns]
crosstab_data = crosstab_data.reindex(columns=actual_col_order)
crosstab_data.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(10, 6))
plt.title("Age Group vs Ad Frustration")
plt.xlabel("Age Group")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.legend(title='Frustration Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 26. Screen Time vs Ad Frustration
Scatter plot to identify correlation.


```
plt.figure()
sns.stripplot(data=df, x='screen_time', y='ads_cause_frustration_score', jitter=True, alpha=0.5, palette='Dark2')
plt.title("Screen Time vs Ad Frustration Score")
plt.xlabel("Screen Time")
plt.ylabel("Frustration Score (1-5)")
plt.xticks(rotation=45)
plt.show()
```

### 27. Occupation vs Ad Fatigue
Box plot analyzing fatigue levels by occupation.


```
plt.figure()
sns.boxplot(data=df, x='occupation', y='ad_fatigue_index', palette='Set1')
plt.title("Occupation vs Ad Fatigue Index")
plt.xlabel("Occupation")
plt.ylabel("Ad Fatigue Index")
plt.xticks(rotation=45)
plt.show()
```

### 28. Ad Fatigue Index Distribution
Histogram of fatigue scores.


```
plt.figure()
sns.histplot(data=df, x='ad_fatigue_index', bins=15, kde=True, color='purple')
plt.title("Ad Fatigue Index Distribution")
plt.xlabel("Ad Fatigue Index")
plt.ylabel("Frequency")
plt.show()
```

### 29. Age Group vs Ad Fatigue Index
Box plot showing fatigue by age group.


```
plt.figure()
sns.boxplot(data=df, x='age_group', y='ad_fatigue_index', palette='muted')
plt.title("Age Group vs Ad Fatigue Index")
plt.xlabel("Age Group")
plt.ylabel("Ad Fatigue Index")
plt.xticks(rotation=45)
plt.show()
```

### 30. Screen Time vs Ad Fatigue Index
Scatter plot analyzing fatigue vs usage time.


```
plt.figure()
sns.stripplot(data=df, x='screen_time', y='ad_fatigue_index', jitter=True, alpha=0.6, palette='Set2')
plt.title("Screen Time vs Ad Fatigue Index")
plt.xlabel("Screen Time")
plt.ylabel("Ad Fatigue Index")
plt.xticks(rotation=45)
plt.show()
```
