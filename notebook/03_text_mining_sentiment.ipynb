```python
import pandas as pd
from textblob import TextBlob
```

```python
df = pd.read_csv("../data/processed/reviews_cleaned.csv")

print("Dataset shape:", df.shape)

df.head()
```

```python
df["clean_text"] = df["clean_text"].fillna("")
```

```python
AD_KEYWORDS = [

    "ads",
    "ad",
    "advertisement",
    "advertisements",
    "advertising",

    "too many ads",
    "too much ads",
    "ads everywhere",
    "ads all the time",
    "ads every time",
    "ads every minute",
    "ads after every",
    "ads after each",

    "video ads",
    "unskippable ads",
    "skippable ads",

    "popup ads",
    "pop up ads",
    "pop-up ads",

    "banner ads",
    "banner ad",

    "interstitial ads",
    "fullscreen ads",

    "annoying ads",
    "hate ads",
    "ads ruining",
    "ads annoying",
    "ads interrupt",
    "ads interrupting",

    "remove ads",
    "premium remove ads",

    "rewarded ads"
]
```

```python
def detect_ads(text):

    text = str(text).lower()

    for word in AD_KEYWORDS:
        if word in text:
            return 1

    return 0
```

```python
def detect_ad_keyword(text):

    text = str(text).lower()

    for word in AD_KEYWORDS:
        if word in text:
            return word

    return "none"
```

```python
def sentiment(text):

    text = str(text)

    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "positive"
    
    elif polarity < 0:
        return "negative"
    
    else:
        return "neutral"
```

```python
df["ad_mentioned"] = df["clean_text"].apply(detect_ads)

df["ad_keyword"] = df["clean_text"].apply(detect_ad_keyword)

df["sentiment"] = df["clean_text"].apply(sentiment)
```
