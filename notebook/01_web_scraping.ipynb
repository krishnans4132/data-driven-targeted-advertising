```python
def scrape_reviews(app_id, app_name, category, country, continent, count=300):

    try:

        result, _ = reviews(
            app_id,
            lang="en",
            country=country,
            sort=Sort.NEWEST,
            count=count
        )

        rows = []

        for r in result:

            rows.append({
                "category": category,
                "app_name": app_name,
                "country": country,
                "continent": continent,
                "review_id": r["reviewId"],
                "review_date": r["at"],
                "rating": r["score"],
                "review_text": r["content"],
                "likes": r["thumbsUpCount"],
                "reviewer_name": r["userName"],
                "scraped_at": datetime.now()
            })

        return pd.DataFrame(rows)

    except Exception as e:
        print("Error:", app_name, country)
        return pd.DataFrame()
```

```python
all_reviews = []

for category, apps in app_categories.items():

    print("Category:", category)

    for app_name, app_id in tqdm(apps.items()):

        for country, continent in countries.items():

            df = scrape_reviews(
                app_id,
                app_name,
                category,
                country,
                continent,
                count=300
            )

            all_reviews.append(df)
```

```python
reviews_df = pd.concat(all_reviews, ignore_index=True)

print("Total reviews collected:", len(reviews_df))

reviews_df.head()
```
