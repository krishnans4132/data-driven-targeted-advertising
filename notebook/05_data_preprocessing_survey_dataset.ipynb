```python
df.columns = [
"timestamp",
"age_group",
"occupation",
"gender",
"smartphone_type",
"screen_time",
"daily_app_categories",
"most_used_category",
"ad_frequency",
"frequent_ad_format",
"most_excessive_ad_category",
"ads_interrupt_usage",
"ads_cause_frustration",
"ads_reduce_enjoyment",
"ads_close_app",
"ads_uninstall_app",
"tolerated_ad_format",
"most_annoying_ad_time",
"pay_for_adfree"
]
```

```python
df.columns
```

```python
df = df.drop(columns=["timestamp"])
```

```python
df.isnull().sum()
```

```python
demographic_cols = [
"age_group",
"occupation",
"gender",
"smartphone_type",
"screen_time"
]

for col in demographic_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
usage_cols = [
"daily_app_categories",
"most_used_category"
]

for col in usage_cols:
    df[col].fillna("Unknown", inplace=True)
```

```python
ads_cols = [
"ad_frequency",
"frequent_ad_format",
"most_excessive_ad_category",
"ads_interrupt_usage",
"ads_cause_frustration",
"ads_reduce_enjoyment",
"ads_close_app",
"ads_uninstall_app",
"pay_for_adfree"
]

for col in ads_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df["tolerated_ad_format"].fillna("None / Not Sure", inplace=True)
```

```python
df["most_annoying_ad_time"].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df = df.convert_dtypes()
```

```python
mapping = {
"Strongly disagree":1,
"Disagree":2,
"Neutral":3,
"Agree":4,
"Strongly agree":5
}

fatigue_cols = [
"ads_interrupt_usage",
"ads_cause_frustration",
"ads_reduce_enjoyment",
"ads_close_app"
]

for col in fatigue_cols:
    df[col+"_score"] = df[col].map(mapping)

df["ad_fatigue_index"] = df[[c+"_score" for c in fatigue_cols]].mean(axis=1)
```
